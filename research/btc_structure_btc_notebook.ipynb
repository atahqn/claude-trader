# BTC Causal Market Structure

Bu notebook, `research.btc_structure` modülünü kullanarak son 5 yıllık BTC verisi üstünde causal `HH/HL/LH/LL`, `EQH/EQL`, `BOS` ve `CHoCH` yapısını üretir ve notebook benzeri interaktif plotlar gösterir.

```python
from pathlib import Path

import pandas as pd

from research.btc_structure import (
    BtcStructureConfig,
    build_global_major_structure_figure,
    build_structure_figure,
    plot_structure_last_n_bars,
    run_btc_structure_pipeline,
)
```

```python
config = BtcStructureConfig.for_interval(
    "1d",
    ticker="BTC/USDT",
    years=5,
)
output_dir = Path("artifacts") / "btc_structure_btcusdt_1d_5y"
artifacts = run_btc_structure_pipeline(config, output_dir=output_dir)
summary = pd.Series(artifacts.summary)
summary
```

```python
summary[[
    "ticker",
    "interval",
    "bars",
    "start",
    "end",
    "confirmed_highs",
    "confirmed_lows",
    "avg_bars_to_confirm_high",
    "avg_bars_to_confirm_low",
    "structure_break_counts",
    "high_label_counts",
    "low_label_counts",
    "last_market_bias_after_close",
]]
```

```python
recent_fig = plot_structure_last_n_bars(
    artifacts,
    last_n=320,
    title="BTC causal structure - last 320 bars",
)
recent_fig
```

```python
full_fig = build_structure_figure(
    artifacts,
    title="BTC causal structure - full 5 year view",
    include_candidates=False,
)
full_fig
```
